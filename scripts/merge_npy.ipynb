```python
import numpy as np
import pandas as pd
import glob
import time

def save_hd5(df, out_name):
    '''
    save into hd5 format for fast reading / writing, and low hard-disk footprint
    :param df:
    :param out_name:
    :return:
    '''
    tic = time.time()
    print('saving: ', df.shape)
    df.to_hdf(out_name, key='null', mode='w', complevel=9, complib='blosc')
    toc = time.time()
    print("saving" + out_name + " took {:.1f} seconds".format(toc - tic))
```

```python
npfiles = glob.glob("./*/*.npy")
npfiles.sort()

i = 0
merged = np.empty((0, 10006))
for idx, npfile in enumerate(npfiles):
    if 'label' not in npfile:
        i += 1 
        arr = np.load(npfile)
        print(npfile, arr.shape)
        merged =  np.append(merged, arr, axis=0)

print("{} npy files merged".format(i))
print("merged arr shape:", merged.shape)
```

```python
merged = pd.DataFrame(merged)
save_hd5(merged, 'merged.hd5')
```
