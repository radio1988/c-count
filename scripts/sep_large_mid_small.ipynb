```python
# load
db_name = '../test/24_15x15.npy'
image_flat_crops = load_blobs_db(db_name)
```

## Filter blobs based on size

### Histogram

```python
r_ = image_flat_crops[:,2]
plt.hist(r_, 40)
plt.title("Histogram of blob size")
plt.show()
```

```python
## Filtering Params
num_shown = 2
min_mid = 30  # mid is target
min_large = 60
```

```python
## small_blobs
small_crops = filter_blobs(image_flat_crops, 0, min_mid)
print('there are {} small blobs'.format(len(small_crops)))

show_rand_crops(crops=small_crops, 
                label_filter=-1, num_shown=num_shown)
```

```python
## mid_blobs
mid_crops = filter_blobs(image_flat_crops, min_mid, min_large)
print('there are {} mid blobs'.format(len(mid_crops)))
show_rand_crops(crops=mid_crops, 
                label_filter=-1, num_shown=num_shown)
```

```python
## large_blobs
large_crops = filter_blobs(image_flat_crops, min_large, 10000000)
print('there are {} large blobs'.format(len(large_crops)))
show_rand_crops(crops=large_crops, 
                label_filter=-1, num_shown=num_shown)
```

```python
np.save('small.npy', small_crops)
np.save('mid.npy', mid_blobs)
np.save('large.npy', large_blobs)
```
